```python
from pylab import *
from numpy import loadtxt
%matplotlib inline
data = loadtxt("sunspots.txt",float)
x = data[:,0]
y = data[:,1]
plot(x,y,label='Noisy')
xlabel("Month")
ylabel("Sunspots")
legend()
show()
```

```python
x = data[:1000,0]
y = data[:1000,1]
xlabel("Month")
ylabel("Sunspots")
plot(x,y,label="Noisy")
legend()
```

```python
L = 1000
r = 5
x = data[:L,0]
y = data[:L,1]
s = [ sum(y[k-r:k+r]) for k in range(L)]
a = array(s)/2/r
plot(x,y,label='Noisy')
plot(x,a,label='Averaged')
xlabel("Month")
ylabel("Sunspots")
legend()
```
