```python
from pylab import plot,show,xlabel,ylabel
from numpy import array,empty,arange,linspace

def ln(r):
    z = 1/2
    for i in range(1001):
	    z = r*z*(1-z)
    return z

def ln1(r):
    a = ln(r)
    for i in range(1001):
	    a = r*a*(1-a)
    return a

def ln2(r):
    b = ln1(r)
    for i in range(1001):
	    b = r*b*(1-b)
    return b

def ln3(r):
    c = ln2(r)
    for i in range(1001):
	    c = r*c*(1-c)
    return c
```

```python
x = linspace(1,4,1000)
y = ln(x)
z = ln1(x)
k = ln2(x)
j = ln3(x)
plot(x,y,"k.")
plot(x,z,"k.")
plot(x,k,"k.")
plot(x,j,"k.")
xlabel("r")
ylabel("x")
```
